```python
items = []
for url in tqdm(urls):
    purl = "https://www.flipkart.com" + (url)
    #print(purl)
    res = get_res(purl)
#     print(res)
    text = res.content
    sel = Selector(res)
    
    jtext = sel.xpath('//*/script[@id="is_script"]/text()').extract()
    jtext = sel.xpath('//*/script[@id="is_script"]/text()').extract()[0]
    jtext = jtext.split("window.__INITIAL_STATE__ = ")[1].replace("};","}").strip() 
    jfile = json.loads(jtext)
    #print(jfile)
    
    
    with open("dells_flipkart.json", "w+") as f:# create file in write mode
        f.write(json.dumps(jfile))
        f.close()
    
    with open('dells_flipkart.json') as access_json: #import and access the data file 
        read_content = json.load(access_json)
        
    item = {}
    item['product_url'] = purl
    
    title = nested_lookup('title', nested_lookup('titles',nested_lookup('pageContext',read_content)))[0]               
    sub_title = nested_lookup('subtitle', nested_lookup('titles',nested_lookup('pageContext',read_content)))[0]  
    item['title_name']  = "".join(title +"(" + (sub_title) + ")").strip()
    
    item['MRP'] = nested_lookup('mrp',nested_lookup('pricing',nested_lookup('pageContext',read_content)))
    item['offer_price'] = nested_lookup('price',nested_lookup('offers',nested_lookup('schema',read_content)))
    
    item['rating_values'] = nested_lookup('ratingValue',nested_lookup('aggregateRating',nested_lookup('schema',read_content)))
    item['review_count'] = nested_lookup('reviewCount',nested_lookup('schema',read_content))
    
    item['sellerName'] = sellerName = nested_lookup('sellerName',read_content)
    item['sellerId'] = sellerName = nested_lookup('sellerId',nested_lookup('trackingDataV2',read_content))
    
    item['Manufactured_by'] = nested_lookup('callouts',nested_lookup('detailedComponents',read_content))[0]
    
    item['Product_Description'] = nested_lookup('description',nested_lookup('schema',read_content))
    item['Sophisticated_Design'] = nested_lookup('text',nested_lookup('description',nested_lookup('featureSetList',read_content)))
  
    media_url ={}
    media = nested_lookup('url',nested_lookup('images',nested_lookup('media',nested_lookup('parentProduct',read_content))))
    for  idx,img_url in enumerate(media):
        img_urls = [img_url.replace("{@width}/{@height}/","").replace("?q={@quality}","").strip()]

        i_url ='img_url{}'
        nimg_url = [i_url.format(idx)]
    #     print(nimg_url)
    #     print(img_urls)        
        for url_key,img_url in zip(nimg_url,img_urls):
            media_url[url_key] = "".join(img_url)
    #     print(media_url)        

    
    Specifications_name = nested_lookup('name',nested_lookup('attributes',read_content))
    Specifications_values =nested_lookup('values',nested_lookup('attributes',read_content))
    Specification_json = {} # to combaine two lists
    for Specifications_nam, Specifications_value in zip(Specifications_name, Specifications_values):
        Specification_json[Specifications_nam] = ", ".join(Specifications_value)



     
    offer_tags = nested_lookup('tags',nested_lookup('offers',nested_lookup('offerGroups',read_content)))
    offers = nested_lookup('title',nested_lookup('offers',nested_lookup('offerGroups',read_content)))
    offer_json = {} # to combaine two lists
    for offer_tag, offer in zip(offer_tags, offers):
        offer_json["".join(offer_tag)] = offer.strip()
    
    items.append(item)
    item.update(media_url)
    item.update(Specification_json)
    item.update(offer_json)
    
    
    
```
