```python
items = []
base_url = "https://www.flipkart.com/search?q=dell+laptop&sid=6bo%2Cb5g&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_4_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_4_na_na_na&as-pos=1&as-type=HISTORY&suggestionId=dell+laptop%7CLaptops&requestId=e7c93711-51ef-4f2e-af94-fc98fd1e84a0&page={}"

for page_no in tqdm(range(1,total_pages+1,1)):
    #print(page_no)
    page_url = base_url.format(page_no)  #to set the page url formate put page number in base url .
    #print(page_url)
    #for url in pro_url:
    res = requests.get(page_url)  #send the request to the url and check respnse code
    sel = Selector(res)           # using selector to scrap the page
    
    jtext = sel.xpath('//*/script[@id="is_script"]//text()').extract()[0]
    jtest = jtext.split("window.__INITIAL_STATE__ = ")[1].replace("};","}").strip() #split, replace, strip halp to fetch dict. file.
    
    
    #load the file in json
    json.loads(jtest)
    
    with open("multi_pageDellData.json", "w") as f:
        f.write(json.dumps(json.loads(jtest)))
    f.close()
    
    
    with open("multi_pageDelldata.json") as access_json:
        jread_content = json.load(access_json)
    
    #set path
    pro_urls = nested_lookup('smartUrl',nested_lookup('productInfo',jread_content))
    img_urls = sel.xpath('//*/img[@class="_396cs4 _3exPp9"]/@src').extract()
    titles = nested_lookup('title',nested_lookup('titles',jread_content))
    offer_prices = nested_lookup('value',nested_lookup('finalPrice',nested_lookup('pricing',jread_content)))
    mrps= nested_lookup('value',nested_lookup('mrp',nested_lookup('pricing',jread_content))) 

    pro_info1 = sel.xpath('//*/ul[@class="_1xgFaf"]/li[1]/text()').extract()
    pro_info2 = sel.xpath('//*/ul[@class="_1xgFaf"]/li[2]/text()').extract()
    pro_info3 = sel.xpath('//*/ul[@class="_1xgFaf"]/li[3]/text()').extract()
    pro_info4 = sel.xpath('//*/ul[@class="_1xgFaf"]/li[4]/text()').extract()
    pro_info5 = sel.xpath('//*/ul[@class="_1xgFaf"]/li[5]/text()').extract()
    pro_info6 = sel.xpath('//*/ul[@class="_1xgFaf"]/li[6]/text()').extract()
    
    
      
    
    for pro_url,img_url,title,offer_price,mrp,pro_info1,pro_info2,pro_info3,pro_info4,pro_info5,pro_info6 in zip(pro_urls,img_urls,titles,offer_prices,mrps,pro_info1,pro_info2,pro_info3,pro_info4,pro_info5,pro_info6):
        
        item = {}
        item['List_Page_url'] = page_url 
        item['product_url'] = pro_url
        item['img_url'] = img_url
        item['product_title'] = title
        item['offer_price'] = offer_price
        item['MRP'] = mrp

        
        item['pro_info1'] = pro_info1
        item['pro_info2'] = pro_info2
        item['pro_info3'] = pro_info3
        item['pro_info4'] = pro_info4
        item['pro_info5'] = pro_info5
        item['pro_info6'] = pro_info6
        
                
        items.append(item)
        
        
    
```

```python
df = pd.DataFrame.from_records(items)
df
```
